```python
sns.barplot(x=df.color,y=df.tip,estimator='mean')
```

The above bar plot shows that the average (mean) tip for the yellow taxis is higher than for the green taxis.

```python
sns.barplot(x=df.color,y=df.tip,estimator='std')
```

The above bar plot shows that yellow taxis shows more variation in tips(higher standard deviation) than the green taxis, yellow tips are less consistent.

```python
sns.histplot(x=df.tip,kde=True,hue=df.color,palette={'yellow':'yellow', 'green':'green'})
```

From above histplot we can say most tips are small in both groups. The yellow taxis has higher and more varied tips, while the green taxiss tips are lower and more consistent.






**Hypotheses:**  
- H0: distribution of tips for yellow taxi = distribution of tips for green taxi  
- H1: distribution of tips for yellow taxi  distribution of tips for green taxi


```python
yellow_tip = df['tip'][df['color'] == 'yellow']
green_tip = df['tip'][df['color'] == 'green']
```

```python
yellow_tip.shape
```

```python
green_tip.shape
```

```python
# Independent T-Test
from scipy import stats

t_stat, p_val = stats.ttest_ind(yellow_tip,green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

#alpha=0.05(significance level)
if p_val < 0.05:
    print("Reject null hypothesis : The tip for yellow taxis is different from that of green")
else:
    print("Fail to reject null hypothesis")
```

So by above analysis we can say that **There is effect of color on tip column.**

**Hypotheses:**

H0: distribution of tips with cash = distribution of tips with credit card

H1: distribution of tips with cash  distribution of tips with credit card

```python
cash_tip = df['tip'][df['payment'] == 'cash']
credit_tip = df['tip'][df['payment'] == 'credit card']
```

```python
cash_tip.shape
```

```python
credit_tip.shape
```

```python
# Independent T-Test
t_stat, p_val = stats.ttest_ind(cash_tip,credit_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

#alpha=0.05(significance level)
if p_val < 0.05:
    print("Reject null hypothesis : The cash tips are different from credit card tips ")
else:
    print("Fail to reject null hypothesis")
```

So by above analysis we can say that **There is effect of payment mode on tip column.**

**Hypotheses:**

H0: Mean tip is same across all pickup_borough categories

H1: Atleast one category has different distribution of tip

```python
manhattan_tip = df['tip'][df['pickup_borough'] == 'Manhattan']
queens_tip = df['tip'][df['pickup_borough'] == 'Queens']
brooklyn_tip = df['tip'][df['pickup_borough'] == 'Brooklyn']
bronx_tip = df['tip'][df['pickup_borough'] == 'Bronx']
```

```python
# ANOVA Test
f_stat, p_val = stats.f_oneway(manhattan_tip,queens_tip,brooklyn_tip,bronx_tip)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Mean tip is diff across all pick_borough")
else:
    print("Fail to reject null hypothesis")
```

**Hypotheses for Tukey test (for each pair of groups):**

H0: distribution of group1 = distribution of group2

H1: distribution of group1  distribution of group2

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey_result = pairwise_tukeyhsd(endog=df.tip, groups=df.pickup_borough, alpha=0.05)
print(tukey_result)
```

```python
# ANOVA Test
#dropoffzone
categories = [group['tip'].values for name, group in df.groupby('dropoff_zone')]
f_stat, p_val = stats.f_oneway(*categories)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

**For pickup_zone**

Count Plot: The plot shows the distribution of tips across various pickup zones, with some zones having many more records than others.

ANOVA Test: The ANOVA test checks if there's a significant difference in tip amounts between the zones. If the p-value is less than 0.05, you reject the null hypothesis, meaning the tips differ across zones.

**For dropoff_zone**

Count Plot: The plot shows the distribution of tips across dropoff zones, with some zones having many more records than others.

ANOVA Test: The ANOVA test shows a p-value approx 0, meaning you reject the null hypothesis, indicating that tip amounts significantly differ across dropoff zones.


So from above analysis we can say that **tip is effected by both pickup_zone and dropoff_zone**

**Hypothesis testing for numerical variables:**

H0 : correlation is zero

H1 : there is significant correlation(not zero)




```python
from scipy.stats import pearsonr,spearmanr
#pearsonr for normal data
#spearmanr for outliers skew data
```

```python
r,p_val=stats.spearmanr(df['tip'],df['passengers'])
print("Correlation :",r)
print("p-value : ",p_val)
if p_val<0.05:
  print("Reject null hypothesis : There is no relationship")
else:
  print("Fail to reject null hypothesis : There is significant relationship")
```
